```python
import numpy as np
from scipy.special import comb
import plotly.graph_objs as go
import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
#  n 
def left(n):
    lr = []
    for i in range(n+1):
        lr.append(comb(n,i))
    lr = np.array(lr)
    return lr/lr.sum()

# 
def change_example(n,v):
    trace1 = go.Bar(
            x=np.arange(n+1),
            y=left(n),visible = v, name = str(n))
    return [trace1]

#  len(s) 
data = []
steps = []
# 
num_particles=[4,5,6,7,15,30,50,100,200,500,700,1000]

for i,n in enumerate(num_particles):
    if n !=num_particles[0]:
        v = False
    else:
        v = True
    data+=change_example(n,v)
    step = dict(
        method = 'restyle',  
        args = ['visible', [False] * len(num_particles)], # 
        label=n
    )
    step['args'][1][i] = True #  i 
    steps.append(step)

sliders = [dict(currentvalue = {'font': {'size': 20},
                                'prefix': 'Total number of particles: '},
                steps = steps)]

        
layout = dict(sliders=sliders,
    xaxis={
        'title':'number of particles on left'
    },
    yaxis={
        'title': 'probability'
    },)
fig = dict(data=data, layout=layout)
plotly.offline.iplot(fig)

#py.iplot(fig, filename='distribution of particles')
```
